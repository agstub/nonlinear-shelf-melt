Here, we outline a simple viscoelastic modification of the viscous contact
problem in the repository: https://github.com/agstub/grounding-line-methods.
Momentum balance and incompressibility are given by
\begin{eqnarray}
-\nabla \cdot \mathbf{\sigma}(\mathbf{u},p) &&= \rho_i\mathbf{g}  \\
\nabla \cdot \mathbf{u} &&= 0. 
\end{eqnarray}
The stress is related to pressure and velocity through
\begin{eqnarray}
\mathbf{\sigma}(\mathbf{u},p)&&= -p\mathsf{I} + \mathbf{\tau},
\end{eqnarray}
where $\mathbf{\tau}$ is the deviatoric stress tensor.
The Maxwell viscoelastic model (see note below) states
\begin{equation}
\mathbf{\tau} + \frac{\eta}{G} \frac{\partial \mathbf{\tau}}{\partial t} = 2\eta\mathsf{D}(\mathbf{u}) \label{maxwell}
\end{equation}
where
\begin{eqnarray}
\mathsf{D}(\mathbf{u}) &&= \frac{1}{2}\left[\nabla \mathbf{u} + (\nabla \mathbf{u})^\mathrm{T}  \right] 
\end{eqnarray}
is the strain rate and $\eta / G$ is the relaxation time
($G$ is the shear modulus and $\eta$ is viscosity).
We make the finite-difference approximation
\begin{equation}
\frac{\partial \mathbf{\tau}}{\partial t}(\mathbf{x},t)
\approx \frac{\mathbf{\tau}(\mathbf{x},t)-\mathbf{\tau}(\mathbf{x},t-\Delta t) }{\Delta t}. 
\end{equation}
$\textbf{Linear viscous case.}$
Defining ${\lambda} = \eta/(G\Delta t)$, we can solve
for $\mathbf{\tau}:=\mathbf{\tau}(\pmb{x},t)$:
\begin{equation}
\mathbf{\tau} =  \frac{2\eta}{1+{\lambda}}\mathsf{D}(\mathbf{u})
+ \frac{\lambda}{1+\lambda} \mathbf{\tau}_0 
\end{equation}
where $\mathbf{\tau}_{0}:=\mathbf{\tau}(\pmb{x},t-\Delta t)$
is known from the previous time step.

Modifications to the weak forms only amount to changing
\begin{equation}
\int_{\Omega} 2\eta\mathsf{D}(\mathbf{u})\mathbf{:}\mathsf{D}(\mathbf{v}) \; \mathrm{d}\mathbf{x}
\quad \mathrm{to} \quad
\int_{\Omega} \frac{2\eta}{1+\lambda}\mathsf{D}(\mathbf{u})\mathbf{:}\mathsf{D}(\mathbf{v}) \; \mathrm{d}\mathbf{x}
+\int_{\Omega} \frac{\lambda}{1+\lambda}\mathbf{\tau}_0\mathbf{:}\mathsf{D}(\mathbf{v}) \; \mathrm{d}\mathbf{x}.
\end{equation}

$\textbf{Nonlinear viscous case...}$ is tricky.
The deviatoric stress evolution equation is really
\begin{eqnarray}
&&\mathbf{\tau} + \frac{\eta(\mathbf{\tau})}{G} \frac{\partial \mathbf{\tau}}{\partial t} = 2\eta(\mathbf{\tau})\mathsf{D}(\mathbf{u}) \\
&&\eta(\mathbf{\tau}) = \frac{1}{2A|\mathbf{\tau}|^{n-1}}
\end{eqnarray}
You can't really solve for $\mathbf{\tau}$ explicitly as above.
However, this type of problem could presumably be solved with a mixed finite element method.

$\textbf{Note on derivation.}$
The Maxwell model states that the elastic/viscous deviatoric stresses
are equal and that the elastic/viscous strains are additive:
\begin{eqnarray}
&&\mathbf{\tau} = 2\eta\dot{\mathbf{\varepsilon}}^\mathrm{viscous}\\
&&\mathbf{\tau} = 2G{\mathbf{\varepsilon}^\mathrm{elastic}} \\
&&\mathbf{\varepsilon} =\mathbf{\varepsilon}^\mathrm{elastic} + \mathbf{\varepsilon}^\mathrm{viscous}.
\end{eqnarray}
The deviatoric stress evolution equation above can be derived from these relations.
